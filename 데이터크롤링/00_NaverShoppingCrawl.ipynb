```python
import pandas as pd
import numpy
import seaborn 
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

# selenium ; MACRO

driver = webdriver.Chrome()
driver.get("https://shopping.naver.com/fresh/directfarm/home")
time.sleep(5)

total = []
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    
    for i in driver.find_elements_by_class_name("_1OKV3UfxuI"):
        title = i.find_element_by_class_name("_1lxNNH5gfD").text
        price = i.find_element_by_class_name("_3MzrqSUlwR").text
        try:
            a = i.find_element_by_class_name("_3G5BbVc9PL.a0tLQiDclf")
            a2 = a.find_element_by_class_name("_29gNnX1yLx").text
            a3 = a2.replace(",","")
            a4 = a3.replace("+","")
            review = int(a4)
            
            b = i.find_element_by_class_name("_3G5BbVc9PL._1FOtzXHiH3")
            c = b.find_element_by_class_name("_29gNnX1yLx").text
            rating = c.split("""\n""")[1][:-2]
        except:
            review = None
            rating = None
        if [title, price, review, rating] not in total:
            total.append([title, price, review, rating])

    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    new_height = driver.execute_script("return document.body.scrollHeight")
    
    if last_height == new_height:
        break
    else:
        last_height = new_height
  
```

```python
a = pd.DataFrame(total, columns = ['제목', "가격", "리뷰", '평점'])
a
```
